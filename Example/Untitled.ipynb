```python
import cobra
```

```python
orig_model = cobra.io.read_sbml_model("model.xml")
new_model = cobra.io.read_sbml_model("model_MCC.xml")
```

```python
for metabolite in new_model.metabolites:
    if any([x < 0 for x in metabolite.elements.values()]):
        print(metabolite.id)
```

```python
for reaction in orig_model.reactions:
    orig_metabolites = {m.id: coeff for m, coeff in reaction.metabolites.items()}
    new_reaction = new_model.reactions.get_by_id(reaction.id)
    new_metabolites = {m.id: coeff for m, coeff in new_reaction.metabolites.items()}
    diff_dict = {}
    for m_id in set(orig_metabolites.keys()).union(new_metabolites.keys()):
        orig_coeff = orig_metabolites.get(m_id, 0)
        new_coeff = new_metabolites.get(m_id, 0)
        diff = orig_coeff - new_coeff
        if diff != 0:
            diff_dict[m_id] = diff
    if len(diff_dict):
        print(f"Reaction {reaction.id} had different coefficients for: {diff_dict}")
```
